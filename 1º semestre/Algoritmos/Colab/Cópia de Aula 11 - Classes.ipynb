#Classes  em Python

Classes permitem agregar dados e funcionalidades, ou seja,  criar novos tipos de dados. 

Exemplo:


```python
#definindo uma classe
class Car:
  brand = None
  model = None
  year = None
```

```python
#criando um objeto (instância) da classe Car
mycar1 = Car()  

mycar1.brand = 'Peugeot'
mycar1.model = 308
mycar1.year = 2015

print("brand: ", mycar1.brand)
print("model:", mycar1.model)
print("year:", mycar1.year)
```

```python
#criando outra instância da classe Car
mycar2 = Car()  

mycar2.brand='chevrolet'
mycar2.model='opala'
mycar2.year='1978'

print("brand: ", mycar2.brand)
print("model:", mycar2.model)
print("year:", mycar2.year)
```

#Criando uma lista com estruturas

```python
class Conta:
  numero = None
  titular = None
  saldo = 0

vetor = []
for i in range(2):
  c = Conta()
  c.numero = input("\nNúmero: ")
  c.titular = input("Titular: ")
  c.saldo = input("Saldo: ")
  vetor.append(c)

#exibindo todas as contas
print("\nRelatório de contas:")
for i in range(2):
  print(" id: ", vetor[i].numero, " cliente: ", vetor[i].titular, " saldo: ", vetor[i].saldo)

#exibindo uma conta específica
print("\nConta armazenada no índice 1:")
print(f'O cliente {vetor[1].titular} possui saldo de R$ {vetor[1].saldo}')
```

#Extra: Inicializando valores na classe 

É possível inicializar os valores das variáveis da classe no momento da instanciação. Para isso, definimos dentro da classe uma função com nome padrão de `__init__`. Esta função recebe os valores a serem usados na inicialização e atribui cada um à respectiva variável da classe. 

A palavra reservada `self` é  usada para mostrar quais variáveis pertencem a classe.

```python
class Car():
  def __init__(self, model, color, company, speed_limit):
    self.color = color
    self.company = company
    self.speed_limit = speed_limit
    self.model = model
```
