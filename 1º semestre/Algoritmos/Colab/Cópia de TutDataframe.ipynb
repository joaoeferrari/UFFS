# Tutorial Pandas Dataframe

```python
import pandas as pd
df=pd.DataFrame()
print(df)
```

Ao instanciar um objeto DataFrame. o construtor possui os seguintes parâmetros:
* data: dados a serem armazenados no dataframe: array, map, dict, outro df, etc


```python
data=[2,4,5,8]
df=pd.DataFrame(data)
print(df)
```

Perceba que as colunas e as linhas não estão rotuladas. O parâmetro **columns** permite criar os rótulos para as colunas

```python
data = [['Banana',3.49],['Laranja',2.20],['Maça',6.5]]
df = pd.DataFrame(data,columns=['Fruta','Preço'])
print(df)
```

Pode-se, também, nomear os linhas (rows/indexes). Normalmente, deixa-se o padrão: 0..n

Parâmetro **index**

```python
data = [['Banana',3.49],['Laranja',2.20],['Maça',6.5]]
df = pd.DataFrame(data,columns=['Fruta','Preço'],index=['Primeiro','Segundo','Terceiro'])
print(df)
```

```python
print(df.columns)
print(df.index)
```

O tipo de dado dicionário (dict) é o mais indicado para criar-se um dataframe com valores pré-definidos

```python
# {'nome coluna':[valores]}
data={'Fruta':['Banana','Laranja','Maçã','Melão','Uva','Limão'],'Preço':[3.49,2.20,6.5,4.3,8.9,5.6]}
df=pd.DataFrame(data)
print(df)
```

Acesso às colunas pode ser feito indexando entre colchetes ou como propriedade

O acesso como propriedade só pode ser utilizado se o nome da coluna não *ferir* uma *palavra reservada* do dataframe

```python
print(df.Fruta)
print(df['Fruta'])
df.Preço
```

```python
data2={'Fruta':['Caqui','Pêssego','Cereja'],'Preço':[1.49,3.20,7.5],'Desconto':['3%','4%','2%']}
df2=pd.DataFrame(data2)
print(df2)
df = pd.concat([df,df2],ignore_index=True) #rearranja os índices (False default usa os índices originais)
df
```

```python
#update (usando o df2)
#método at
df2.at[0,'Fruta']='Kaki'
print("at")
print(df2)
#método loc
print("loc linha específica")
df2.loc[0,['Fruta']]='Caqui'
print(df2)
df2.loc[0,['Fruta','Preço','Desconto']]=['Caqui',10,'10%']
print("loc linha específica e multi coluna")
print(df2)
df2.loc[0:1,['Fruta','Preço','Desconto']]=['Caqui',10,'10%']
print("loc multi linha e multi coluna")
print(df2)
```

```python
download_url = ("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
df=pd.read_csv(download_url)
print(df.columns)
df.head()
df.groupby('Major_category').agg({'Men':['mean','std'],'Women':['mean','std']})
#df.plot(x="Rank", y=["P25th", "Median", "P75th"])
df['Median'].plot.hist()
top5=df.sort_values(by='Median', ascending=False).head()
top5.plot.bar(x='Major',y='Median',rot=45) # rot -> rotação em graus / padrão 90
medians6000 = df[df["Median"] > 60000].sort_values("Median")
medians6000.plot.bar(x='Major',y=["P25th", "Median", "P75th"])
```

```python
Tcat=df.groupby("Major_category")["Total"].sum().sort_values()
for k, v in Tcat.items():
  print(k,v)
Tcat.plot.barh()
```
